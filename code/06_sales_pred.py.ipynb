```python
import pandas as pd
from sklearn.linear_model import LinearRegression

df = pd.read_csv("../data/cleaned_data.csv")

# 每日销量
daily_sales = df[df['behavior']=='buy'].groupby('date').size().reset_index()
daily_sales.columns = ['date', 'sales']
daily_sales['day'] = range(len(daily_sales))
daily_sales['date'] = pd.to_datetime(daily_sales['date'])
# 假设 date 列已经是 datetime 类型（清洗时已处理），直接使用 .dt.weekday
daily_sales['is_weekend'] = (daily_sales['date'].dt.weekday >= 5).astype(int)

X = daily_sales[['day', 'is_weekend']]
y = daily_sales['sales']
model = LinearRegression()
model.fit(X, y)

# 预测下一天
last_date = daily_sales['date'].max()
next_date = last_date + pd.Timedelta(days=1)
next_day_num = len(daily_sales)
next_is_weekend = 1 if next_date.weekday() >= 5 else 0
next_day_df = pd.DataFrame({'day': [next_day_num], 'is_weekend': [next_is_weekend]})
pred = model.predict(next_day_df)

print(f"✅ 考虑周末效应的线性回归预测：次日（{next_date.date()}）销量≈{int(pred[0])}单")
daily_sales.to_csv("../result/sales_pred.csv", index=False)
```
