```python
import pandas as pd
df = pd.read_csv("../data/cleaned_data.csv")
```

```python
# ============== 1. 用户行为漏斗分析（电商核心转化率） ==============
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']  # 解决中文显示
plt.rcParams['axes.unicode_minus'] = False
behavior_cnt = df['behavior'].value_counts()
pv = behavior_cnt['pv']  # 点击
cart = behavior_cnt['cart']  # 加购
fav = behavior_cnt['fav']  # 收藏
buy = behavior_cnt['buy']  # 支付

print("===== 电商转化漏斗 =====")
print(f"点击量：{pv}")
print(f"加购+收藏：{cart+fav}")
print(f"支付量：{buy}")
print(f"整体转化率：{round(buy/pv*100,2)}%")

# ============== 2. 热销商品TOP10 ==============
hot_item = df[df['behavior']=='buy'].groupby('item_id').size().sort_values(ascending=False).head(10)
print("\n===== 热销商品TOP10 =====")
print(hot_item)

# ============== 3. 每日访客数趋势 ==============
daily_user = df.groupby('date')['user_id'].nunique()
daily_user.plot(kind='line', title='每日独立访客数', figsize=(10,4))
plt.savefig("../result/daily_user.png")
plt.close()

# 保存结果
result = pd.DataFrame({
    '指标':['点击量','加购收藏','支付量','转化率'],
    '数值':[pv,cart+fav,buy,round(buy/pv*100,2)]
})
result.to_csv("../result/basic_report.csv",index=False)
print("✅ 基础业务报表完成！")
```
