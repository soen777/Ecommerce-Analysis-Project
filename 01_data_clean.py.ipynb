```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
# 1. 读取数据
df = pd.read_csv('../data/UserBehavior.csv',
                 names=['user_id', 'item_id', 'category_id', 'behavior', 'timestamp'])
```

```python
print("原始数据量：", len(df))

# 2. 缺失值处理
print("缺失值统计：\n", df.isnull().sum())
df = df.dropna()

# 3. 重复值处理
df = df.drop_duplicates()
print("去重后数据量：", len(df))

# 4. 时间戳转正常时间（
df['time'] = pd.to_datetime(df['timestamp'], unit='s')
df['date'] = df['time'].dt.date
df['hour'] = df['time'].dt.hour

# 5. 异常值处理
# 电商数据时间范围：2017-11-25 至 2017-12-03、
start_date = pd.to_datetime('2017-11-25')
end_date = pd.to_datetime('2017-12-04')
df = df[(df['time'] >= start_date) & (df['time'] < end_date)]

print("清洗后最终数据量：", len(df))

# 6. 保存清洗后数据
df.to_csv("../data/cleaned_data.csv", index=False, encoding='utf-8-sig')
print("✅ 数据清洗完成！文件已保存到 data/cleaned_data.csv")
```
